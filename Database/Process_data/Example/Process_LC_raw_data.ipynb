```python
# Get foreground data: red band
df_frg_r = df_frg[df_frg["filter"]=='r'] 
df_frg_r
```

## Plot the data

```python
fig, ax = plt.subplots(figsize=([8,4]))
plt.scatter(df_frg_g['MJD'],df_frg_g['unforced_mag'],color= 'green',label="g", alpha = 0.7)
plt.scatter(df_frg_r['MJD'],df_frg_r['unforced_mag'],color= 'red',label="r", alpha = 0.7)
plt.scatter(df_bkg_g['MJD'],df_bkg_g['unforced_mag'],color= 'green',label="g_bkg",marker= "^", alpha = 0.3)
plt.scatter(df_bkg_r['MJD'],df_bkg_r['unforced_mag'],color= 'red',label="r_bkg", marker= "^",alpha = 0.3)
plt.ylabel('Magnitude')
plt.gca().invert_yaxis()
plt.xlabel('Days')
plt.title(SN_name)
plt.minorticks_on()
plt.grid(which='minor', color='#999999', linestyle='-', alpha=0.2)
plt.legend()
plt.legend(fontsize = "small", loc = 1)
plt.show()
```
